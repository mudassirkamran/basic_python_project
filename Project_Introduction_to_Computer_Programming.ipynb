##ğŸ“¦ğŸ“¬ğŸšš Creating a shipping cost calculator! âš™ï¸

You need to create a program that calculates the cost of shipping a package based on its weight and destination ğŸ“¦ğŸ’°

ğŸŒ Domestic shipping within the country :

* Packages weighing up to 1 kg cost Rs.50 to ship ğŸ¤‘
Each additional kg costs Rs.10 extra ğŸ’°

ğŸŒ International shipping to other countries ğŸŒ:

* Packages weighing up to 1 kg cost Rs.500 to ship ğŸ¤‘

Each additional kg costs Rs.100 extra ğŸ’°
Ready to ship? ğŸš›ğŸš€

```python
#Calulating the shipping cost of the package.
wt = int(input('Please enter the weight of the package in kg: '))
mode = input('Please enter the mode - domestic/international: ').lower()

# Domestic

if mode == 'domestic':
  base_cost = 50
  if wt <= 1:
    domestic_cost = base_cost
  else:
    domestic_cost = base_cost + (10 * (wt-1))
  print(f'The cost to ship {wt} kg of package will be {domestic_cost} rs.')

if mode == 'international':
  base_cost = 500
  if wt <= 1:
    international_cost = base_cost
  else:
    international_cost = base_cost + (100*(wt-1))
  print(f'The cost to ship {wt} kg of package will be {international_cost} rs.')
```

## You're given a nested list of numbers, and your task is to count the number of odd and even numbers in the list. To achieve this, you'll use a nested loop and some basic conditional statements.

    [1, 6, 21, 18, 3],
    [10, 13, 25, 33, 30],

]

```python
# List of numbers
numbers = [
    [2, 5, 11, 20, 8],
    [9, 4, 15, 28, 17],
    [1, 6, 21, 18, 3],
    [10, 13, 25, 33, 30],
    [14, 7, 16, 19, 22]
]

even_count = 0
odd_count = 0

for x in numbers:
  for y in x:
    if y % 2 == 0:
      even_count += 1
    else:
      odd_count += 1

print('Even Count: ', even_count)
print('Odd Count: ', odd_count)
```

## Imagine you're organizing a chess tournament ğŸ†, where each player has to play against every other player. To achieve this, you can use nested loops in Python! Nested loops are simply loops within loops, and they come in handy when you need to perform tasks involving multiple levels of iteration.

In Python, you can nest 'for' loops, 'while' loops, or even a combination of both. Let's take a closer look at how nested loops work using our chess tournament scenario:

```python
# Write Your Code Here
players = ['Alice', 'Bob', 'Charlie', 'Diana']
for x in players:
  for y in players:
    if x != y:
      print(f"{x} will play with {y}")
```

In this activity, you will build a simple calculator using conditional looping in Python. The calculator will ask the user for two numbers and an operation (addition, subtraction, multiplication, or division), and then perform the operation on the two numbers.

```python
# Write code here
first_num = int(input("Please enter the first number: "))
second_num = int(input("Please enter the second number: "))
op = input("What do you want to perform +,-,*,/ : ")

if op == '+':
  print(first_num + second_num)
elif op == '-':
  print(first_num - second_num)
elif op == '*':
  print(first_num * second_num)
elif op == '/':
  if second_num != 0:
    print(first_num/second_num)
  else:
    print("Cannot divide the number with 0!")
```

## **Building a Guessing Game**

In this activity, you will build a simple guessing game using conditional looping in Python. The game will generate a random number between 1 and 100, and the user will have to guess the number within a certain number of tries. The program will provide hints to the user after each guess, telling them whether their guess is too high or too low.

```python
import random
secret_num = random.randint(1,100)

chance = 10

while chance != 0:
  guessed_num = int(input("Please guess the number: "))
  diff = abs(guessed_num - secret_num)

  if guessed_num == secret_num:
    print("You guessed the number correct")
    break

# Difference checks
  if diff > 50:
          print("There is a difference of 50+")
  elif 40 < diff <= 50:
          print("There is a difference of 40â€“50")
  elif 30 < diff <= 40:
          print("There is a difference of 30â€“40")
  elif 20 < diff <= 30:
          print("There is a difference of 20â€“30")
  elif 10 < diff <= 20:
          print("There is a difference of 10â€“20")
  elif 5 < diff <= 10:
          print("There is a difference of 5â€“10")
  elif 2 < diff <= 5:
          print("There is a difference of 2â€“5")
  elif diff <= 2:
          print("You are very close!")

  chance -= 1

  if chance == 0:
    print("You ran out of chances!! The number was:", secret_num)
```

Example:

```python
# Write code here

start = int(input("Enter the first number: "))
end = int(input("Enter the second number: "))

digit = int(input("Enter digit to count: "))
count = 0
for x in range(start,end+1):
  if str(x%10) == digit:
    count += 1
  x //= 10
print(count)


```

## **Lambda Functions**

ğŸ§‘â€ğŸ’» You are working as a Python developer for a startup company that has just received a project from a client. The client wants a program that can perform certain operations on lists using lambda functions. Your manager has assigned you the task of finding the sum of squared odd numbers from the list.

ğŸ‘‰ Hint: You have to use filter, map and reduce functions to get the task done.

```python
# Import reduce from functools module
from functools import reduce

# Create a list of numbers from 1 to 10
numbers = list(range(1, 11))

odd_num = list(filter(lambda x: x% 2 != 0, numbers))
squared_odds = list(map(lambda x:x**2,odd_num))
result = reduce(lambda a, b: a + b, squared_odds)
print(result)
```

## **What's the time?** âŒšâŒš

Create a class Time which takes two inputs: hours and minutes to instantiate.

* Construct a method DisplayTime() which displays the time in AM/PM formats.

 * For example if the input is 14 hours and 45 mins, then this method will print "The time is 2:45 PM".
 * If the inputted hours exceeds 23 then print the message "The input hours should be less than 24" and if the inputted minutes exceeds 59 then print the message "The input minutes should be less than 60."
 * Also if the input is 12 hours 30 minutes, then the displayed time would be 12:30 PM

* Construct a method DisplayRatio() which should display the ratio of minutes to hours.
 * For example, (8 hours and 16 mins) should display 2. Use try, except block to account for ZeroDivisionError.

```python
class Time:
    def __init__(self):
        # Hour input
        while True:
            try:
                self.hour = int(input("Please enter the hour time (1 to 12): "))
                if 1 <= self.hour <= 12:
                    break
                else:
                    print("Hour must be between 1 and 12.")
            except:
                print("Please enter a valid number.")

        # Minute input
        while True:
            try:
                self.minutes = int(input("Please enter the minute time (0 to 59): "))
                if 0 <= self.minutes <= 59:
                    break
                else:
                    print("Minutes must be between 0 and 59.")
            except:
                print("Please enter a valid number.")

        # Period input
        while True:
            self.period = input("Please enter AM or PM: ").strip().upper()
            if self.period in ["AM", "PM"]:
                break
            else:
                print("Please enter a valid period (AM or PM).")

    def current_time(self):
        print(f"The current time is {self.hour}:{self.minutes:02d} {self.period}")
```
